```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data.xlsx - Sheet1.csv')

# Displaying the first few rows
print(df.head())
print(df.shape)
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['Salary'], kde=True)
plt.title('Salary Distribution')
plt.show()
```

Observation: This code creates a histogram of the 'Salary' column with a density estimate showing a  right skewed salary distribution with a peak indicating frequent salary range.

```python
sns.boxplot(x=df['Salary'])
plt.title('Boxplot of Salary')
plt.show()
```

Observation: The boxplot of 'Salary' shows the median, quartiles, and outliers, showing a right-skewed distribution with outliers.

```python
sns.scatterplot(x='collegeGPA', y='Salary', data=df)
plt.title('Salary vs College GPA')
plt.show()
```

Observation: Scatter plot shows positive correlation between college GPA and salary.

```python
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('Salary by Gender')
plt.show()
```

Observation: boxplot shows salary distribution by gender with visible differences

```python
# Average salary by job title
avg_salary_by_designation = df.groupby('Designation')['Salary'].mean()
print(avg_salary_by_designation)
```

```python
# Gender and Specialization analysis
gender_specialization = pd.crosstab(df['Gender'], df['Specialization'])
gender_specialization.plot(kind='bar', stacked=True)
plt.title('Specialization Preference by Gender')
plt.show()
```

Observation: more males prefer technical specializations compared to females

```python
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Scatter plot of Salary vs College GPA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='collegeGPA', y='Salary', data=df)
plt.title('Salary vs College GPA')
plt.xlabel('College GPA')
plt.ylabel('Salary')
plt.show()

# Linear regression analysis
X = df['collegeGPA']
y = df['Salary']
X = sm.add_constant(X)  # Adds a constant term to the predictor

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Plotting the regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='collegeGPA', y='Salary', data=df)
plt.plot(df['collegeGPA'], predictions, color='red')  # Regression line
plt.title('Salary vs College GPA with Regression Line')
plt.xlabel('College GPA')
plt.ylabel('Salary')
plt.show()

# Displaying regression results
print(model.summary())
```

Observation: higher college GPA tends to correlate with higher salary
